```python
import pandas as pd
vehicles = pd.read_csv('vehicles.csv')
vehicles.head()
```

1. We've created a Boolean filter `is_ford_truck` that returns `True` whenever a vehicle is a Ford pickup truck. Create a new Boolean filter `not_ford_truck` that is `True` for every vehicle that is not a Ford pickup truck.

```python
is_ford = vehicles['manufacturer'] == 'ford'
is_truck = vehicles['vehicle_type'] == 'pickup'
is_ford_truck = is_ford & is_truck


## YOUR SOLUTION HERE ##
not_ford_truck = ~is_ford_truck
```

2. Use `not_ford_truck` to filter `vehicles` to only those vehicles that aren't Ford pickup trucks. Assign the result to `not_ford_truck_vehicles`.

```python
## YOUR SOLUTION HERE ##
not_ford_truck_vehicles = vehicles[not_ford_truck]

# show output
not_ford_truck_vehicles.head()
```

3. Filter `vehicles` down to only vehicles manufactured by someone **other** than `chevrolet` or `gmc`. Assign the result to `not_chevrolet_gmc`.

```python
## YOUR SOLUTION HERE ##
is_chev = vehicles['manufacturer'] == 'chevrolet'
is_gmc = vehicles['manufacturer'] == 'gmc'
chev_gmc = is_chev | is_gmc


not_chevrolet_gmc = ~chev_gmc
not_chevrolet_gmc = vehicles[not_chevrolet_gmc]

# show output
not_chevrolet_gmc.head()
```
