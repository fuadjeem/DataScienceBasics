```python
import pandas as pd
parks = pd.read_csv('nationalpark_visitors.csv')
parks.head()
```

1. Rename the columns `Year2021`, `Year2020`, and `Year2019` to `Visitors2021`, `Visitors2020`, and `Visitors2019`, respectively.

```python
## YOUR SOLUTION HERE ##
column_mapper = {
    'Year2021' : 'Visitors2021',
    'Year2020' : 'Visitors2020',
    'Year2019' : 'Visitors2019',
}
parks = parks.rename(mapper = column_mapper, axis = 1)

# show output
parks.head()
```

2. Drop the `index` column from `parks`

```python
## YOUR SOLUTION HERE ##
drop_columns = ['index']
parks = parks.drop(labels = drop_columns, axis = 1)

# show output
parks.head()
```
