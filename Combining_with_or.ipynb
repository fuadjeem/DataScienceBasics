```python
import pandas as pd
vehicles = pd.read_csv('vehicles.csv')
vehicles.head()
```

1. Combine the comparisons `is_ethanol` and `is_methanol` to produce `True` for any row whose fuel is either ethanol or methanol. Assign the result to `combined`.

```python
is_ethanol = vehicles['fuel'] == 'ethanol (e85)'
is_methanol = vehicles['fuel'] == 'methanol'
## YOUR SOLUTION HERE ##
combined = is_ethanol | is_methanol
```

2. Use `combined` to filter `vehicles` down to only those rows containing vehicles running on either ethanol or methanol. Assign the result to `ol_vehicles`.

```python
## YOUR SOLUTION HERE ##
ol_vehicles = vehicles[combined]

# show output
ol_vehicles.head()
```

3. Filter `vehicles` down to only those rows with `fuel` equal to `electric` or `best_mpge` greater than `40` (the average.) Assign the result to `electric_or_above_average`.

```python
## YOUR SOLUTION HERE ##
is_electric = vehicles['fuel'] == 'electric'
is_best = vehicles['best_mpge'] > 40

electric_or_above_average = vehicles[is_electric | is_best]

# show output
electric_or_above_average.head()
```
