```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
#playing with colors
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'solid')
plt.plot(x, np.sin(x+1), color = 'red', linestyle = 'dashed')
plt.plot(x, np.sin(x+2), color = 'green', linestyle = 'dotted')
plt.plot(x, np.sin(x+3), color = '#FFDD44', linestyle = 'dashdot')
""" plt.plot(x, np.cos(x), color = 'blue')
plt.plot(x, np.cos(x+1), color = 'red')
plt.plot(x, np.cos(x+2), color = 'green')
plt.plot(x, np.cos(x+3), color = '#FFDD44') """
plt.show()
```

```python
#playing with colors
plt.plot(x, np.sin(x),'-g')
plt.plot(x, np.sin(x+1),'--c')
plt.plot(x, np.sin(x+2),'-.k')
plt.plot(x, np.sin(x+3),':r')
plt.show()
```

```python
# to limit the x and y axis and giving a title to the axis

plt.plot(x, np.sin(x),'-g', label = 'sin(x)')
plt.plot(x, np.sin(x+1),'--c', label = 'sin(x+1)')
plt.xlim(-1,12)
plt.ylim(-5,12)
#plt.xlabel('sin(x)')
#plt.ylabel('sin(x+1)')
plt.title('This is the title for the plot')
plt.legend()
plt.show
```
