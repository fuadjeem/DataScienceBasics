```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
#playing with colors
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'solid')
plt.plot(x, np.sin(x+1), color = 'red', linestyle = 'dashed')
plt.plot(x, np.sin(x+2), color = 'green', linestyle = 'dotted')
plt.plot(x, np.sin(x+3), color = '#FFDD44', linestyle = 'dashdot')
""" plt.plot(x, np.cos(x), color = 'blue')
plt.plot(x, np.cos(x+1), color = 'red')
plt.plot(x, np.cos(x+2), color = 'green')
plt.plot(x, np.cos(x+3), color = '#FFDD44') """
plt.show()
```

```python
#playing with colors
plt.plot(x, np.sin(x),'-g')
plt.plot(x, np.sin(x+1),'--c')
plt.plot(x, np.sin(x+2),'-.k')
plt.plot(x, np.sin(x+3),':r')
plt.show()
```

```python
# to limit the x and y axis and giving a title to the axis

plt.plot(x, np.sin(x),'-g', label = 'sin(x)')
plt.plot(x, np.sin(x+1),'--c', label = 'sin(x+1)')
plt.xlim(-1,12)
plt.ylim(-5,12)
#plt.xlabel('sin(x)')
#plt.ylabel('sin(x+1)')
plt.title('This is the title for the plot')
plt.legend()
plt.show
```

```python
import pandas as pd

sp_x = [2,3,2,3,6,6,5,5,4,6,7,8,1,6]
sp_y = [6,8,7,9,4,6,5,4,6,5,6,2,6,1]
df = pd.DataFrame({'A':sp_x, 'B':sp_y})
df.plot(kind='scatter', x ='A', y = 'B')
plt.show()
```

```python
df.plot(kind='bar', x ='A', y = 'B')
plt.show()
```

```python
lx = [7,1,4,5,9,2,3]
ly = [98,2,32,158,15,25,18]
df = pd.DataFrame({'A': lx,'B' : ly})
df
```
